```python
import pandas as pd
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['ps.useafm']=True
matplotlib.rcParams['pdf.use14corefonts']=True
matplotlib.rcParams['text.usetex']=True
```

```python
yelp = pd.read_excel("./feature%.xlsx").iloc[:10]
yelp
```

```python
amazon = pd.read_excel("./feature%.xlsx").iloc[11:]
amazon
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))
ax = axes.flatten()

pd.DataFrame(amazon).plot(ax=ax[0], style=['g^-', 'ro-', 'k*-', 'y-', 'bs-'])
ax[0].set_xlabel("number of features", fontsize=25, fontweight='heavy')
ax[0].set_ylabel("Amazon NDCG@50", fontsize=15, fontweight='heavy')
ax[0].legend(loc='lower right')

pd.DataFrame(yelp).plot(ax=ax[1], style=['g^-', 'ro-', 'k*-', 'y-', 'bs-'])
ax[1].set_xlabel("number of features", fontsize=25, fontweight='heavy')
ax[1].set_ylabel("Yelp NDCG@50", fontsize=15, fontweight='heavy')
ax[1].legend(loc='lower right')

fig.tight_layout()
plt.savefig("./feature_compare.pdf", format='pdf')
```
