```python
import pandas as pd
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['ps.useafm']=True
matplotlib.rcParams['pdf.use14corefonts']=True
matplotlib.rcParams['text.usetex']=True
```

```python
yelp = pd.read_excel("./BPR.xlsx").iloc[:13]
```

```python
ax = pd.DataFrame(yelp).plot(style=['ro-', 'bs-', 'g^-'])
ax.set_xlabel("BPR weight")
ax.set_ylabel("Yelp NDCG@50")
```

```python
amazon = pd.read_excel("./BPR.xlsx").iloc[16:]
amazon
```

```python
ax = pd.DataFrame(amazon).plot(style=['ro-', 'bs-', 'g^-'])
ax.set_xlabel("BPR weight")
ax.set_ylabel("Amazon NDCG@50")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(9, 3))
ax = axes.flatten()

pd.DataFrame(amazon).plot(ax=ax[0], style=['ro-', 'bs-', 'g^-'], linewidth=2)
ax[0].set_xlabel("BPR weight", fontsize=15, fontweight='heavy')
ax[0].set_ylabel("Amazon NDCG@50", fontsize=12, fontweight='heavy')
ax[0].legend(loc='lower right')

pd.DataFrame(yelp).plot(ax=ax[1], style=['ro-', 'bs-', 'g^-'], linewidth=2)
ax[1].set_xlabel("BPR weight", fontsize=15, fontweight='heavy')
ax[1].set_ylabel("Yelp NDCG@50", fontsize=12, fontweight='heavy')
ax[1].legend(loc='lower right')

fig.tight_layout()
plt.savefig("./BPR_compare.pdf", format='pdf')
```
