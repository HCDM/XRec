```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
matplotlib.rcParams['ps.useafm']=True
matplotlib.rcParams['pdf.use14corefonts']=True
matplotlib.rcParams['text.usetex']=True
```

```python
amazon = pd.read_excel("./max_depth.xlsx").iloc[:3]
amazon
```

```python
yelp = pd.read_excel("./max_depth.xlsx").iloc[6:9]
yelp
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(9, 3))
ax = axes.flatten()

pd.DataFrame(amazon).plot(ax=ax[0], style=['ro-', 'bs-', 'g^-'], linewidth=2)
ax[0].set_xlabel("depth", fontsize=20, fontweight='heavy')
ax[0].set_ylabel("Amazon NDCG@50", fontsize=15, fontweight='heavy')
ax[0].legend(loc='center right')
ax[0].set_xticks([4 ,5, 6])



pd.DataFrame(yelp).plot(ax=ax[1], style=['ro-', 'bs-', 'g^-'], linewidth=2)
ax[1].set_xlabel("depth", fontsize=20, fontweight='heavy')
ax[1].set_ylabel("Yelp NDCG@50", fontsize=15, fontweight='heavy')
ax[1].legend(loc='center right')
ax[1].set_xticks([4 ,5, 6])


# plt.suptitle("Varying the depth of user tree", y=1.05)


fig.tight_layout()
plt.savefig("./depth_compare_user_tree.pdf", format='pdf')
```
